```python
import pandas as pd
import glob
import joblib
```

```python
data_paths= '/data/female-amb/impute/100K'
```

```python
all_cools = glob.glob(f'{data_paths}/*.cool')
all_cell_id = [path.split('/')[-1].split('.')[0] for path in all_cools]
cell_table = pd.DataFrame({'cell_url':all_cools}, index = all_cell_id)
```

```python
use_cts = joblib.load('/data/metadata/m3c_use_cts')
use_cts = [ct.replace(' ','_').replace('/','') for ct in use_cts]

meta = pd.read_csv('/data/metadata/240104_m3C_META.csv', index_col = 0)
meta['AgingMajorType'] = meta['AgingMajorType'].apply(lambda x: x.replace(' ','_').replace('/',''))
meta = meta[meta['AgingMajorType'].isin(use_cts)]
meta['AgingMajorType.Age'] = meta['AgingMajorType'] + '.' + meta['Age']
```

```python
cell_table['AgingMajorType.Age'] = meta['AgingMajorType.Age']
cell_table['AgingMajorType'] = meta['AgingMajorType']
cell_table['Age'] = meta['Age']
```

```python
for path in glob.glob('/data/tmp/cool_tables/*.tsv'):
    ct_age = path.split('/')[-1].split('.tsv')[0]
    tmpdf_25k_cells = pd.read_csv(path, sep = '\t', header = None)[0].tolist()
    print(f"{ct_age}:{len(tmpdf_25k_cells)} cells")
    tmpdf_100k = cell_table.loc[tmpdf_25k_cells]
    tmpdf_100k['cell_id'] = tmpdf_100k.index
    tmpdf_100k = tmpdf_100k[['cell_url','cell_id']]
    tmpdf_100k.to_csv(f"100k_cool_tables/{ct_age}.tsv", sep = '\t', header = None, index = None)    
```
