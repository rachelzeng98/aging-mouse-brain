```python
import karyopype.karyopype as kp
import pandas as pd
import matplotlib.pyplot as plt
from collections import defaultdict
plt.rcParams['figure.dpi'] = 100
plt.rcParams['savefig.dpi']=300
plt.rcParams['pdf.fonttype']=42
```

```python
_dir = '500_result'
_type = 'Hyper'
```

```python
#tmp = pd.read_csv('/home/qzeng/pkg/karyopype/tests/data/testRegions/regions.bed', header = None)
```

```python
bed_path = f'{_dir}/{_type}.hotspot.bed'
hotspot_bed = pd.read_csv(bed_path, sep = '\t', header = None)

re_bed = defaultdict(list)
for _, row in hotspot_bed.iterrows():
    _chr, start, end = row[0], row[1], row[2]
    re_bed[0].append(f"{_chr} {start} {end}")
bed_df = pd.DataFrame(re_bed)
bed_df.to_csv(f"{_dir}/kp_{_type}.bed", index = None, header = None)
```
