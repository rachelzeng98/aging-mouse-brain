```python
#_ct = 'OPC_NN'
_type = 'Hypo'
known_motifs = glob.glob(f"re_condition_dmr/*/motif_*.{_type}//knownResults.txt")
len(known_motifs)
```

```python
tf_oi = ['NFkB-p50,p52(RHD)', 'NFkB-p65(RHD)', 'NFkB-p65-Rel(RHD)','CTCF(Zf)',]
```

```python
all_df =[]

for motif_result in known_motifs:
    result_df = pd.read_csv(motif_result, sep="\t", index_col=0)
    result_df.index = [x.split("/")[0] for x in result_df.index]
    cell_type =  motif_result.split('/')[-3]
    tmpdf = result_df.loc[result_df.index.isin(tf_oi)]
    tmpdf['celltype'] = cell_type
    all_df.append(tmpdf)
    
all_df = pd.concat(all_df, axis = 0)
```
