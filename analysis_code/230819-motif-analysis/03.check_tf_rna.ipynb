```python
tf_table= pd.read_csv(f"/home/qzeng/ref/mouse/allTFs_mm.gene_info.csv", index_col = 0)
tf_table.head()
```

```python
dir_path = '/home/qzeng/project/aging/240518_redo_aDMG/DEGs'

all_df =[]
for fp in tqdm(glob.glob(f'{dir_path}/*')):
    ct = fp.split('/')[-1].split('.')[0]
    tmpdf = pd.read = pd.read_csv(fp, index_col =0)
    tmpdf['log2(aged/young)'] = -tmpdf['avg_log2FC']
    tmpdf['-log(p)'] = -np.log(tmpdf['p_val'] + 1e-50)
    tmpdf['celltype'] = ct
    shared_genes = list(set(tf_table['gene_name']) & set(tmpdf.index)) 
    tmpdf = tmpdf.loc[shared_genes]
    all_df.append(tmpdf)
    
all_df = pd.concat(all_df, axis = 0)
all_df.head()
```

```python
all_df = all_df[all_df['p_val'] < 0.05]
```

```python
gene_names = [
    "Fra1", 
    "Batf", 
    "Fosl2", 
    "AP-1", 
    "Fos", 
    "JunB", 
    "Jun-AP1", 
    "Atf3", 
    "Fra2",
    'Jun'
]
```
