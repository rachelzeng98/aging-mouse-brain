```python
import anndata
import matplotlib.pyplot as plt
import pandas as pd
from ALLCools.plot import *
from amb import *
import joblib
from ALLCools.mcds import MCDS
```

```python
def get_gene_values_by_name(gene_name):
    data = gene_frac_da.sel(geneslop2k=gene_name_to_gene_id[gene_name]).to_pandas()
    data.name = gene_name
    return data
```

```python
_dir = '/home/qzeng/project/aging/231117_update_annot'
coord_base = 'tsne'
mcds_path = '/ceph/gale-1/qzeng/AmbData/geneslop2k_frac.mcds'
mc_type = 'CGN'

gene_annotation_path = '/gale/netapp/home/wtian/refs/mouse/mm10/gencode/v22/gencode.vM22.basic.annotation.gene.flat.tsv.gz'
mc_meta_path = '/home/qzeng/project/aging/231117_update_annot/240104_mC_META.csv'
```

```python
adata = anndata.read_h5ad(f'{_dir}/final_with_coords.h5ad')
mc_meta = pd.read_csv(mc_meta_path, index_col = 0)
adata.obs['Class'] = mc_meta['Class']
```

```python
mc_data = adata[adata.obs["Modality"] == "mC"]
neu_mc_data = adata[(adata.obs["Modality"] == "mC") & (~adata.obs["Class"].isin(['NN', 'IMN']))]
nn_mc_data = adata[(adata.obs["Modality"] == "mC") & (adata.obs["Class"].isin(['NN', 'IMN']))]
```

```python
gene_frac_da = MCDS.open(
    mcds_path, 
    obs_dim='cell', 
    #var_dim='chrom100k',
    use_obs=mc_data.obs_names,  
)['geneslop2k_da_frac']
gene_frac_da
```

```python
gene_frac_da = gene_frac_da.sel(mc_type=mc_type).load()
gene_frac_da
```

```python
gene_meta = pd.read_csv(gene_annotation_path, sep = '\t')
gene_name_to_gene_id = gene_meta.set_index('gene_name')['gene_id'].to_dict()
gene_id_to_gene_name = gene_meta.set_index('gene_id')['gene_name'].to_dict()
```

```python
cell_num = mc_meta['AgingMajorType'].value_counts()
use_cts = cell_num.head(30).index
```

```python
ct = 'OPC NN'
gene_name = marker_gene_dict[ct]
```

```python
fig, ax = plt.subplots(figsize=(4,4), dpi=300)

categorical_scatter(data=mc_data,
                        coord_base=coord_base,
                        max_points=None,
                        hue=None,
                        rasterized=True,
                        scatter_kws=dict(color='lightgrey'),
                        ax=ax)

_ = continuous_scatter(ax=ax,
                       data=nn_mc_data,
                       #hue = gene_name,
                       hue=get_gene_values_by_name(gene_name),
                       rasterized=True,
                       coord_base=coord_base,
                       max_points=None,
                       )
ax.set_title(gene_name)

fig.savefig(f'{gene_name}.{ct}.pdf', bbox_inches = 'tight')
```
