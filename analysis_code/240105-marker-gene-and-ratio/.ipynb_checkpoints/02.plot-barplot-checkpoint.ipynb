```python
import pandas as pd
from amb import *
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mt_order = ['DG Glut',
 'CA1-ProS Glut',
 'CA3 Glut',
 'L2/3 IT CTX Glut',
 'L4/5 IT CTX Glut',
 'L6 IT CTX Glut',
 'L5 ET CTX Glut',
 'L5 NP CTX Glut',
 'L2/3 IT PPP Glut',
 'L2/3 IT ENT Glut',
 'L6 CT CTX Glut',
 'L6b/CT ENT Glut',
 'LA-BLA-BMA-PA Glut',
 'PAG Glut',
 'STR D1 Gaba',
 'STR D2 Gaba',
 'STR-PAL Chst9 Gaba',
 'STR D1 Sema5a Gaba',
 'Sst Gaba',
 'Pvalb Gaba']
```

```python
mc_meta_path = '/home/qzeng/project/aging/231117_update_annot/240104_mC_META.csv'
m3c_meta_path = '/home/qzeng/project/aging/231117_update_annot/240104_m3C_META.csv'

age_order= ['8wk','9mo','18mo']
```

```python
m3c_meta= pd.read_csv(m3c_meta_path, index_col = 0)
```

```python
counts = m3c_meta.groupby("AgingMajorType")["Age"].value_counts().unstack().fillna(0)

ratio = counts/counts.sum()
ratio = ratio.div(ratio.sum(axis = 1), axis = 0)
ratio = ratio[age_order]

cum_sum_data = ratio.cumsum(axis=1)
```

```python
# counts = (m3c_meta.groupby("AgingMajorType")["Age"].value_counts().unstack().fillna(0))
# counts = counts[age_order]

# ratios = counts / counts.sum(axis=1).values[:, None]
# cum_sum_data = ratios.cumsum(axis=1)
```

```python
cum_sum_data = cum_sum_data.loc[mt_order]
```

```python
plot_data = cum_sum_data.unstack().reset_index()
plot_data.columns = [ "Age", "CellType","height"]
```

```python
age_palette = get_age_palette()
age_palette['8wk']  ='#480080'
```

```python
fig, axes = plt.subplots(figsize=(len(mt_order) * 0.5, 4), 
                       ncols = len(mt_order), dpi=200)

for cell_type, ax in zip(mt_order, axes.ravel()):
    use_plot_data = plot_data[plot_data["CellType"] == cell_type]
    
    for i, age in enumerate(age_order[::-1]):
        ct_plot_data = use_plot_data[use_plot_data["Age"] == age]
        if ct_plot_data.shape[0] == 0:
            continue
        sns.barplot(
            ax = ax,
            data=ct_plot_data,
            x="CellType",
            y="height",
            color=age_palette[age],
        )
    
    ax.axhline(y=0.33, color='white', linestyle='--') 
    ax.axhline(y=0.66, color='white', linestyle='--') 
    
    ax.xaxis.set_tick_params(rotation=90)
    ax.set_title(cell_type,  rotation=90, fontsize = 7)
    ax.axis("off")
fig.savefig("AgingMajorType.Age.Ratio.pdf", bbox_inches = 'tight')
```
