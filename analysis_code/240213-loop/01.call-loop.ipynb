```python
from schicluster.loop import call_loop
import pandas as pd
import os
import subprocess
import pathlib
```

```python
group_name = 'Vip_Gaba.9mo'
```

```python
cell_table_path = f"/data/tmp/loop_tables/{group_name}.csv"
```

```python
output_bucket_dir = "ecker-rachel-amb-datasets/female-amb/diff-domain"
```

```python
call_loop(
      cell_table_path=cell_table_path,
      output_dir=group_name,
      chrom_size_path=os.path.expanduser('/ref/m3C/mm10.main.nochrM.nochrY.chrom.sizes'),
      shuffle=True,
      chunk_size=200,
      dist=5050000,
      cap=5,
      pad=5,
      gap=2,
      resolution=10000,
      min_cutoff=1e-06,
      keep_cell_matrix=False,
      cpu_per_job=48,
      log_e=True,
      raw_resolution_str=None,
      downsample_shuffle=None,
      black_list_path=os.path.expanduser('/ref/m3C/mm10_bismark_loop_blacklist.bed'),
      fdr_pad=7,
      fdr_min_dist=5,
      fdr_max_dist=500,
      fdr_thres=0.1,
      dist_thres=20000,
      size_thres=1)
```

```python
subprocess.run(f"gsutil -m cp -r {group_name} gs://{output_bucket_dir}", shell = True)
subprocess.run(f"touch {group_name}/{group_name}.finish", shell = True)
subprocess.run(f'gsutil cp {group_name}/{group_name}.finish gs://{output_bucket_dir}', shell = True)
```
