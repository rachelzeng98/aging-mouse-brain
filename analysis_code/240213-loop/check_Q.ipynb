```python
import cooler
import matplotlib.pyplot as plt
import numpy as np
```

```python
ct_age = 'Oligo_NN.9mo'
chrom = 'chr19'
ll, rr = 8514358, 11514358
```

```python
cool_path = f'/data/female-amb/impute/merged-25k/{ct_age}.Q.cool'
cool = cooler.Cooler(cool_path)
Q = cool.matrix(balance=False, sparse=True).fetch(chrom).tocsr()

resl = 25000
loopl, loopr = (ll//resl), (rr//resl)
print(f"{loopl} {loopr}")
tmp = Q[loopl:loopr, loopl:loopr].toarray()

fig, ax = plt.subplots(figsize = (3,3), dpi = 200)
img = ax.imshow(tmp, cmap='afmhot_r', vmin=0, vmax=0.02, aspect='auto')
```
